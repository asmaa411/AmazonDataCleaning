<h1> Data Cleaning Delivery<h1>

```python
delivery = pd.read_csv(r'D:\Data engineer project.amazon\Datasets\amazon_delivery.csv')
delivery
```

```python
delivery.dtypes
```

```python
delivery=delivery.drop(['Agent_Age', 'Pickup_Time'], axis = 1)
delivery=delivery.dropna()
new_columns_name = {
    'Delivery_Time':'Delivery_Time_Min'
    }

delivery.rename(columns = new_columns_name, inplace = True)
```

```python
delivery
```

```python
from datetime import datetime
delivery['Order_Time'] = delivery['Order_Time'].apply(lambda x: datetime.strptime(x, '%H:%M:%S').time())
```

```python
import folium
```

```python
df_limit = delivery.head(1000)

df_unique = df_limit.drop_duplicates(subset=['Store_Latitude', 'Store_Longitude', 'Drop_Latitude', 'Drop_Longitude'])
map_center = [df_unique['Store_Latitude'].mean(), df_unique['Store_Longitude'].mean()]
my_map = folium.Map(location=map_center, zoom_start=5)
for _, row in df_unique.iterrows():
    source = [row['Store_Latitude'], row['Store_Longitude']]
    destination = [row['Drop_Latitude'], row['Drop_Longitude']]
    folium.Marker(location=source, popup="Store").add_to(my_map)
    folium.Marker(location=destination, popup="Destination").add_to(my_map)
my_map.save("map_with_unique_lines.html")
```

```python
delivery.to_csv('Cleaned_Delivery.csv', index = False, header = True, encoding = 'utf-8', sep = ',', mode = 'w')
```

<h1>Data Cleaning Amazon<h1>

```python
amazon= pd.read_csv(r'D:\Data engineer project.amazon\Datasets\amazon.csv')
amazon
```

```python
amazon.dtypes
```

```python
amazon=amazon.drop(['user_name', 'about_product', 'product_link','img_link','review_content','review_title','review_id'], axis = 1)
amazon=amazon.dropna()
```

```python
new_columns_name = {
    'discounted_price':'discounted_price_INR',
    'actual_price':'actual_price_INR'
    }
amazon.rename(columns = new_columns_name, inplace = True)
```

```python
amazon['discounted_price_INR']=amazon['discounted_price_INR'].str.replace('₹' ,'' ,regex = False).str.strip()
amazon['actual_price_INR']=amazon['actual_price_INR'].str.replace('₹' ,'' ,regex = False).str.strip()
amazon['discount_percentage']=amazon['discount_percentage'].str.replace('%' ,'' ,regex = False).str.strip()
amazon['rating_count']=amazon['rating_count'].str.replace(',' ,'' ,regex = False).str.strip()
amazon['actual_price_INR']=amazon['actual_price_INR'].str.replace(',' ,'' ,regex = False).str.strip()
amazon['discounted_price_INR']=amazon['discounted_price_INR'].str.replace(',' ,'' ,regex = False).str.strip()
```

```python
amazon['discounted_price_INR'] = pd.to_numeric(amazon['discounted_price_INR'], errors = 'coerce')
amazon['actual_price_INR'] = pd.to_numeric(amazon['actual_price_INR'], errors = 'coerce')
amazon['rating']=pd.to_numeric(amazon['rating'] , errors='coerce')
amazon['discount_percentage']=amazon['discount_percentage'].astype('int')
amazon['rating_count']=amazon['rating_count'].astype('int')
```

```python
amazon.dtypes
```

```python
amazon['amount_discounted'] = amazon['actual_price_INR'] - amazon['discounted_price_INR']
column_to_move = 'amount_discounted'
cols = list(amazon.columns)
cols.insert(cols.index('actual_price_INR') + 1, cols.pop(cols.index(column_to_move)))
amazon = amazon[cols]
amazon
```

```python
amazon.to_csv('Cleaned_amazon.csv', index = False, header = True, encoding = 'utf-8', sep = ',', mode = 'w')
```
